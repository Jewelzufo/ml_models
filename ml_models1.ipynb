```python
# Import the necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
X_new = np.array([[6]])
prediction = model.predict(X_new)

# Display the data on a graph
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, model.predict(X), color='red', label='Linear Regression Model')
plt.scatter(X_new, prediction, color='green', label='Prediction')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```

```python
# Import the necessary libraries
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate random data points
X = np.random.rand(100, 2)  # 100 data points with 2 features

# Create a KMeans clustering model
kmeans = KMeans(n_clusters=3)  # Define the number of clusters

# Fit the model to the data
kmeans.fit(X)

# Get cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Display the data and clusters on a graph
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', label='Data Points')
plt.scatter(centers[:, 0], centers[:, 1], marker='x', color='red', label='Cluster Centers')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering Model')
plt.legend()
plt.show()
```
